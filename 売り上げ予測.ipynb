```python
# datetimeを分割して設定
train['year'] = train['datetime'].apply(lambda x : x.split('-')[0])
train['month'] = train['datetime'].apply(lambda x : x.split('-')[1])
train['day'] = train['datetime'].apply(lambda x : x.split('-')[2])
test['day'] = test['datetime'].apply(lambda x : x.split('-')[2])
test['year'] = test['datetime'].apply(lambda x : x.split('-')[0])
test['month'] = test['datetime'].apply(lambda x : x.split('-')[1])

# 新規で定義したカラムのデータ型はobject型になってしまうため、int型へ変換する
train['year'] = train['year'].astype(np.int)
train['month'] = train['month'].astype(np.int)
train['day'] = train['day'].astype(np.int)
test['day'] = test['day'].astype(np.int)
test['year'] = test['year'].astype(np.int)
test['month'] = test['month'].astype(np.int)
```

```python
train=train.fillna(0)
train
```

```python
x_train = pd.get_dummies(train[["week","weather", "temperature"]])
x_train.head()
```

```python
def remark(x):
  if x==0:
    return 0
  else:
    return 1
```

```python
def remark1(x):
  if x=="月":
    return 0
  elif x=="火":
    return 1
  elif x=="水":
    return 2
  elif x=="木":
    return 3
  else:
    return 4
```

```python
def remark2(x):
  if x=="快晴":
    return 0
  elif x=="晴れ":
    return 1
  elif x=="薄雲":
    return 2
  elif x=="曇り":
    return 3
  elif x=="雨":
    return 4
  else:
    return 5
```

```python
train['event1'] = train['event'].apply(lambda x : remark(x))
test['event1'] = test['event'].apply(lambda x : remark(x))
train['week1'] = train['week'].apply(lambda x : remark1(x))
test['week1'] = test['week'].apply(lambda x : remark1(x))
train['weather1'] = train['weather'].apply(lambda x : remark2(x))
test['weather1'] = test['weather'].apply(lambda x : remark2(x))
```

```python
from sklearn.neural_network import MLPClassifier
from sklearn.naive_bayes import CategoricalNB
ALPHA = 1e-10
x_train = train[["month","week1", "weather1","temperature","event1"]]
x_test = test[["month","week1", "weather1","temperature","event1"]]
y_train = train["y"]
#model = LR()
model = CategoricalNB(alpha=ALPHA)
#model = MLPClassifier(solver="sgd", random_state=0, max_iter=3000)
model.fit(x_train, y_train)
pred = model.predict(x_test)
pred
```
